```python
import pandas as pd
import numpy as np
```

```python
es_files = ['ES_2017.csv',
            'ES_2018.csv',
            'ES_2019.csv',
            'ES_2020.csv',
            'ES_2021.csv',
            'ES_2022.csv',
            'ES_2023.csv']
ym_files = ['YM_2017.csv',
            'YM_2018.csv',
            'YM_2019.csv',
            'YM_2020.csv',
            'YM_2021.csv',
            'YM_2022.csv',
            'YM_2023.csv']

columns = ['date', 'time', 'open', 'high', 'low', 'close', 'volume']
```

```python
df_es = pd.DataFrame(columns=columns)
for i, file in enumerate(es_files):
    if i == 0:
        df_es = pd.read_csv(f'data/{file}', names=columns, dtype={'date': 'str', 'time': 'str'})
    else:
        df_es = pd.concat([df_es, pd.read_csv(f'data/{file}', names=columns, dtype={'date': 'str', 'time': 'str'})])


df_es['date'] = '20' + df_es['date'].str[:2] + '-' + df_es['date'].str[2:4] + '-' + df_es['date'].str[4:]
df_es['time'] = df_es['time'].str[:2] + ':' + df_es['time'].str[2:4] + ':' + df_es['time'].str[4:]
df_es['date'] = pd.to_datetime(df_es['date'] + ' ' + df_es['time']).dt.tz_localize('America/New_York')
df_es = df_es.drop(columns=['time'])
df_es = df_es.set_index('date')


df_ym = pd.DataFrame(columns=columns)
for i, file in enumerate(ym_files):
    if i == 0:
        df_ym = pd.read_csv(f'data/{file}', names=columns, dtype={'date': 'str', 'time': 'str'})
    else:
        df_ym = pd.concat([df_ym, pd.read_csv(f'data/{file}', names=columns, dtype={'date': 'str', 'time': 'str'})])

df_ym['date'] = '20' + df_ym['date'].str[:2] + '-' + df_ym['date'].str[2:4] + '-' + df_ym['date'].str[4:]
df_ym['time'] = df_ym['time'].str[:2] + ':' + df_ym['time'].str[2:4] + ':' + df_ym['time'].str[4:]
df_ym['date'] = pd.to_datetime(df_ym['date'] + ' ' + df_ym['time']).dt.tz_localize('America/New_York')
df_ym = df_ym.drop(columns=['time'])
df_ym = df_ym.set_index('date')

```

```python
df_nq = pd.read_feather('data/NQ_2017_2023.feather')
df_nq.index  = df_nq.index - pd.Timedelta(minutes=1)

df_nq_2023 = pd.read_csv('data/NQ_2023.csv', names=columns, dtype={'date': 'str', 'time': 'str'})
df_nq_2023['date'] = df_nq_2023['date'].str[:4] + '-' + df_nq_2023['date'].str[4:6] + '-' + df_nq_2023['date'].str[6:]
df_nq_2023['time'] = df_nq_2023['time'].str[:2] + ':' + df_nq_2023['time'].str[2:4] + ':' + df_nq_2023['time'].str[4:]
df_nq_2023['date'] = pd.to_datetime(df_nq_2023['date'] + ' ' + df_nq_2023['time']).dt.tz_localize('America/New_York')
df_nq_2023 = df_nq_2023.drop(columns=['time'])
df_nq_2023 = df_nq_2023.set_index('date')
df_nq = pd.concat([df_nq, df_nq_2023])
df_nq.drop_duplicates(inplace=True)
```

```python

df_es.to_feather('data/ES_2017_2023.feather')
df_ym.to_feather('data/YM_2017_2023.feather')
df_nq.to_feather('data/NQ_2017_2023.feather')

```

```python

def count_moves(df, tick_size):
    df['body_size'] = (abs(df['close'] - df['open']) /tick_size).astype(int)
    df['top_wick'] = ((df['high'] - df[['open', 'close']].max(axis=1)) /tick_size).astype(int)
    df['bottom_wick'] = ((df[['open', 'close']].min(axis=1) - df['low']) /tick_size).astype(int)
    df['direction'] = np.sign(df['close'] - df['open']).astype(int)
    df['gap_size'] = (abs(df['open'] - df['close'].shift(1)) /tick_size)
    df['gap_direction'] = np.sign(df['open'] - df['close'].shift(1))
    df.fillna(0, inplace=True)
    df['gap_size'] = df['gap_size'].astype(int)
    df['gap_direction'] = df['gap_direction'].astype(int)
    df['hour_start'] = (df.index.minute == 0).astype(int)
    df['15min_start'] = (((df.index.minute % 15) == 0).astype(int) * (df.index.minute // 15)).isin(range(1,4)).astype(int)
    df['5min_start'] = (((df.index.minute % 5) == 0).astype(int) * (df.index.minute // 5)).isin(range(1,12)).astype(int)
    return df

df_es = count_moves(df_es, 0.25)
df_ym = count_moves(df_ym, 0.25)
df_nq = count_moves(df_nq, 0.25)
```

```python
def create_move(row, tick_size):
    if row['hour_start'] == 1:
        move = 'H'
    elif row['15min_start'] == 1:
        move = 'Q'
    elif row['5min_start'] == 1:
        move = 'F'
    else:
        move = 'M'
        
    if row['gap_direction'] == 1: # if there is gap up after the candle
        move = move + 'G' * int(row['gap_size'])
    elif row['gap_direction'] == -1: # if there is gap down after the candle
        move =  move + 'g' * int(row['gap_size'])

    if row['direction'] == 1 or row['direction'] == 0: # green candle. Assume doji is green
        move =  move + 'D' * int(row['bottom_wick']) + 'U' * int(row['bottom_wick'] + row['body_size'] + row['top_wick']) + 'D' * int(row['top_wick'])
    else: # red candle
        move = move + 'U' * int(row['bottom_wick']) + 'D' * int(row['bottom_wick'] + row['body_size'] + row['top_wick']) + 'U' * int(row['top_wick'])
    
    return move + 'E'

def create_move1(row, tick_size):
    if row['hour_start'] == 1:
        move = 'H'
    elif row['15min_start'] == 1:
        move = 'Q'
    elif row['5min_start'] == 1:
        move = 'F'
    else:
        move = 'M'
        
    if row['direction'] == 1 or row['direction'] == 0: # green candle. Assume doji is green
        move =  move + 'D' + str(int(row['bottom_wick'])) + 'U' + str(int(row['bottom_wick'] + row['body_size'] + row['top_wick'])) + 'D' + str(int(row['top_wick']))
    else: # red candle
        move = move + 'U' + str(int(row['bottom_wick'])) + 'D' + str(int(row['bottom_wick'] + row['body_size'] + row['top_wick'])) + 'U' + str(int(row['top_wick']))
    if row['gap_direction'] == 1: # if there is gap up after the candle
        move = move + 'G' + str(int(row['gap_size']))
    elif row['gap_direction'] == -1: # if there is gap down after the candle
        move =  move + 'g' + str(int(row['gap_size']))
    return move + 'E'


df_es['move'] = df_es.apply(lambda row: create_move(row, 0.25), axis=1)
df_ym['move'] = df_ym.apply(lambda row: create_move(row, 1), axis=1)
df_nq['move'] = df_nq.apply(lambda row: create_move(row, 0.25), axis=1)
```

```python
#join all dataframes by date
df = pd.concat([df_es[['move']], df_ym[['move']], df_nq[['move']]],axis=1, join='inner')
df = df.dropna()
```
